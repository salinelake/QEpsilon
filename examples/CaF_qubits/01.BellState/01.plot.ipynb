```python
from matplotlib import pyplot as plt
import numpy as np
import os

data_folder = '../00.Ramsey/Data'
data_XY8_235 = np.loadtxt(os.path.join(data_folder, 'Fig3E_PurpleTriangles.csv'), delimiter=',', skiprows=1)
data_XY8_235 = data_XY8_235[np.argsort(data_XY8_235[:, 0])]
data_XY8_193 = np.loadtxt(os.path.join(data_folder, 'Fig3E_BlueCircles.csv'), delimiter=',', skiprows=1)
data_XY8_193 = data_XY8_193[np.argsort(data_XY8_193[:, 0])]
data_XY8_168 = np.loadtxt(os.path.join(data_folder, 'Fig3E_GreenSquares.csv'), delimiter=',', skiprows=1)
data_XY8_168 = data_XY8_168[np.argsort(data_XY8_168[:, 0])]
data_XY8_160 = np.loadtxt(os.path.join(data_folder, 'Fig3E_YellowDiamonds.csv'), delimiter=',', skiprows=1)
```

```python
temp=20
preparation_rate = 0.79

tweezer_sep=2.35
t_235 = np.load('T{:.0f}uK_d{:.2f}um_t.npy'.format(temp, tweezer_sep) ) / 1000 ##to ms
P_235 = np.load('T{:.0f}uK_d{:.2f}um_P00.npy'.format(temp, tweezer_sep) )
loss_235 = np.load('T{:.0f}uK_d{:.2f}um_loss.npy'.format(temp, tweezer_sep) )
P_235 *= preparation_rate**2

tweezer_sep=1.93
t_193 = np.load('T{:.0f}uK_d{:.2f}um_t.npy'.format(temp, tweezer_sep) ) / 1000 ##to ms
P_193 = np.load('T{:.0f}uK_d{:.2f}um_P00.npy'.format(temp, tweezer_sep) )
loss_193 = np.load('T{:.0f}uK_d{:.2f}um_loss.npy'.format(temp, tweezer_sep) )
P_193 *= preparation_rate**2

tweezer_sep=1.68
t_168 = np.load('T{:.0f}uK_d{:.2f}um_t.npy'.format(temp, tweezer_sep) ) / 1000 ##to ms
P_168 = np.load('T{:.0f}uK_d{:.2f}um_P00.npy'.format(temp, tweezer_sep) )
loss_168 = np.load('T{:.0f}uK_d{:.2f}um_loss.npy'.format(temp, tweezer_sep) )
P_168 *= preparation_rate**2

## also get those data generated without thermal motion
temp = 0

tweezer_sep=2.35
t_235_motion_free = np.load('T{:.0f}uK_d{:.2f}um_t_motion_free.npy'.format(temp, tweezer_sep) ) / 1000 ##to ms
P_235_motion_free = np.load('T{:.0f}uK_d{:.2f}um_P00_motion_free.npy'.format(temp, tweezer_sep) )
loss_235_motion_free = np.load('T{:.0f}uK_d{:.2f}um_loss_motion_free.npy'.format(temp, tweezer_sep) )
P_235_motion_free *= preparation_rate**2

tweezer_sep=1.93
t_193_motion_free = np.load('T{:.0f}uK_d{:.2f}um_t_motion_free.npy'.format(temp, tweezer_sep) ) / 1000 ##to ms
P_193_motion_free = np.load('T{:.0f}uK_d{:.2f}um_P00_motion_free.npy'.format(temp, tweezer_sep) )
loss_193_motion_free = np.load('T{:.0f}uK_d{:.2f}um_loss_motion_free.npy'.format(temp, tweezer_sep) )
P_193_motion_free *= preparation_rate**2

tweezer_sep=1.68
t_168_motion_free = np.load('T{:.0f}uK_d{:.2f}um_t_motion_free.npy'.format(temp, tweezer_sep) ) / 1000 ##to ms
P_168_motion_free = np.load('T{:.0f}uK_d{:.2f}um_P00_motion_free.npy'.format(temp, tweezer_sep) )
loss_168_motion_free = np.load('T{:.0f}uK_d{:.2f}um_loss_motion_free.npy'.format(temp, tweezer_sep) )
P_168_motion_free *= preparation_rate**2
```

```python
import matplotlib as mpl
mpl.rcParams['axes.linewidth'] = 2
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['lines.markersize'] = 5

fig, ax = plt.subplots(4,1, figsize=(4,4), sharex=True,)
ax[0].plot(t_235, P_235, color='tab:purple', linewidth=2, markersize=0, label=r'DIQCD')
ax[0].plot(data_XY8_235[:,0], data_XY8_235[:,1], marker='o', color='tab:purple', linewidth=0, label=r'EXP')
ax[0].plot(t_235_motion_free, P_235_motion_free, color='tab:purple', alpha=0.3, linewidth=3, markersize=0, label='No motion' )
ax[1].plot(t_193, P_193, color='tab:blue', linewidth=2, markersize=0, label=r'DIQCD')
ax[1].plot(data_XY8_193[:,0], data_XY8_193[:,1], marker='o', color='tab:blue', linewidth=0, label=r'EXP')
ax[1].plot(t_193_motion_free, P_193_motion_free, color='tab:blue', alpha=0.3, linewidth=3, markersize=0, label='No motion'  )
ax[2].plot(t_168, P_168, color='tab:green', linewidth=2, markersize=0, label=r'DIQCD')
ax[2].plot(data_XY8_168[:,0], data_XY8_168[:,1], marker='o', color='tab:green', linewidth=0, label=r'EXP')
ax[2].plot(t_168_motion_free, P_168_motion_free, color='tab:green', alpha=0.3, linewidth=3, markersize=0, label='No motion'  )
ax[3].plot(t_235, loss_235 * 100,  color='tab:purple', linewidth=2, label='2.35 $\mathrm{\mu m}$')
ax[3].plot(t_193, loss_193 * 100,  color='tab:blue', linewidth=2, label='1.93 $\mathrm{\mu m}$')
ax[3].plot(t_168, loss_168 * 100,  color='tab:green', linewidth=2, label='1.68 $\mathrm{\mu m}$')
ax[0].set_ylabel(r'$P_{\upuparrows}$', fontsize=14)
ax[1].set_ylabel(r'$P_{\upuparrows}$', fontsize=14)
ax[2].set_ylabel(r'$P_{\upuparrows}$', fontsize=14)
ax[3].set_ylabel('Loss [%]', fontsize=14)
ax[3].set_xlabel('t [ms]', fontsize=14)
## set tick label by array
ax[3].set_yticks([0,  10,  20])
ax[0].text(0.7, 0.05, r'$d$=2.35 $\mathrm{\mu m}$', fontsize=10, transform=ax[0].transAxes, color='tab:purple')
ax[1].text(0.7, 0.05, r'$d$=1.93 $\mathrm{\mu m}$', fontsize=10, transform=ax[1].transAxes, color='tab:blue')
ax[2].text(0.7, 0.05, r'$d$=1.68 $\mathrm{\mu m}$', fontsize=10, transform=ax[2].transAxes, color='tab:green')

for _ax in ax[:3]:
    _ax.set_ylim(-0.15, 0.75)
    _ax.set_yticks([0, 0.2, 0.4, 0.6])
    _ax.legend(frameon=False, fontsize=10, loc='upper right', ncols=3, columnspacing=0.5, handletextpad=0.3, handlelength=1.0)
ax[3].set_ylim(-1, 24)

plt.tight_layout()
plt.subplots_adjust(hspace=0)
plt.savefig('compare.png',dpi=300)
plt.show()
plt.close()
```
